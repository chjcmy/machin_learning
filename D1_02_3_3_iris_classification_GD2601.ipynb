# 데이터셋 불러오기

### 1. 라이브러리 준비

**1. 이유 (Why):** 요리 도구를 꺼내는 단계입니다. 엑셀 기능을 하는 pandas, 계산기 numpy, 그림 그리는 matplotlib를 가져옵니다.

**2. 중요 포인트 (Key Point):** 가장 많이 쓰는 기본 도구들입니다.

**3. 주목할 점 (Focus):** 에러 없이 실행되면 성공입니다.


```python
# 데이터 분석을 위해 pandas와 numpy 라이브러리를 임포트합니다.
# pandas: 데이터 조작 및 분석
# numpy: 수치 계산
import pandas as pd
import numpy as np
```

### 2. 데이터 가져오기

**1. 이유 (Why):** 분석할 '붓꽃 데이터'를 로드합니다.

**2. 중요 포인트 (Key Point):** sklearn 라이브러리에 내장된 연습용 데이터를 사용합니다.

**3. 주목할 점 (Focus):** iris 변수에 데이터가 담깁니다.


```python
# sklearn 라이브러리에서 데이터셋 모듈을 임포트합니다.
from sklearn import datasets

# load_iris() 함수를 사용하여 붓꽃(Iris) 데이터셋을 로드합니다.
# 이 데이터셋은 딕셔너리와 유사한 Bunch 객체로 반환됩니다.
iris = datasets.load_iris()

# 데이터셋 객체의 키(key)들을 출력하여 구성요소를 확인합니다.
# data, target, frame, target_names, DESCR, feature_names 등이 포함되어 있습니다.
iris.keys()
```

```python
# 'DESCR' 키를 사용하여 데이터셋에 대한 상세한 설명(Description)을 출력합니다.
# 데이터의 특성, 통계 정보, 출처 등을 확인할 수 있습니다.
print(iris['DESCR'])
```

```python
# 'target' 키에 저장된 라벨(정답) 데이터의 크기(shape)를 확인합니다.
print("데이터셋 크기:", iris['target'].shape)

# 'target' 데이터의 내용을 출력합니다. 0, 1, 2 등의 정수로 클래스가 인코딩되어 있습니다.
print("데이터셋 내용: \n", iris['target'])
```

```python
# 'data' 키에 저장된 특성(feature) 데이터의 크기(shape)를 확인합니다.
print("데이터셋 크기:", iris['data'].shape)

# 'data' 데이터의 내용 중 첫 7개 행만 슬라이싱하여 출력합니다.
print("데이터셋 내용: \n", iris['data'][:7, :])
```

### 3. 데이터프레임 변환

**1. 이유 (Why):** 컴퓨터가 좋아하는 형태(Array)를 사람이 보기 편한 표(DateFrame)로 바꿉니다.

**2. 중요 포인트 (Key Point):** 이 'df' 변수가 앞으로 분석의 핵심이 됩니다.

**3. 주목할 점 (Focus):** 표 모양이 예쁘게 잘 나왔는지 확인하세요.


```python
# 'data' 배열을 사용하여 pandas DataFrame을 생성합니다.
# 컬럼 이름은 'feature_names' 키의 값을 사용합니다.
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])

# 생성된 DataFrame의 크기(행과 열의 수)를 확인합니다.
print("데이터프레임의 형태:", df.shape)

# DataFrame의 첫 5개 행을 출력하여 데이터를 미리 봅니다.
df.head()
```

```python
# 컬럼 이름을 코딩하기 편하도록 간결하게 변경합니다.
# 공백과 단위를 제거하고 소문자와 언더바(_)를 사용합니다.
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# 변경된 컬럼 이름이 적용된 DataFrame의 첫 2개 행을 확인합니다.
df.head(2)
```

```python
# 'Target'이라는 새로운 컬럼을 만들고, 여기에 iris['target'] 데이터를 할당합니다.
df['Target'] = iris['target']

# 타겟 컬럼이 추가된 후의 DataFrame 크기를 확인합니다.
print('데이터셋의 크기: ', df.shape)

# 데이터의 첫 5개 행을 출력하여 'Target' 컬럼이 잘 추가되었는지 확인합니다.
df.head()
```

# Baseline 모델 학습

### 5. 학습용/시험용 나누기

**1. 이유 (Why):** 로봇이 문제와 답을 몽땅 외우는(과적합) 것을 막기 위해, 공부할 문제(Train)와 시험 문제(Test)를 나눕니다.

**2. 중요 포인트 (Key Point):** test_size=0.2는 20%를 시험용으로 뺀다는 뜻입니다.

**3. 주목할 점 (Focus):** X_train(공부용), X_test(시험용)의 개수가 잘 나뉘었는지 보세요.


```python
from sklearn.model_selection import train_test_split

X_data = df.loc[:, 'sepal_length':'petal_width']
y_data = df.loc[:, 'Target']

# 데이터셋을 학습용(Train)과 테스트용(Test)으로 분리합니다.
# test_size=0.2: 전체 데이터의 20%를 테스트용으로 사용합니다.
# shuffle=True: 데이터를 무작위로 섞어서 분리합니다.
# random_state=20: 매번 동일한 결과를 얻기 위해 난수 시드를 고정합니다.
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data,
                                                    test_size=0.2,
                                                    shuffle=True,
                                                    random_state=20)

# 분리된 데이터셋의 크기를 확인합니다.
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# KNN(K-Nearest Neighbors) 분류기 모델을 임포트합니다.
from sklearn.neighbors import KNeighborsClassifier

# n_neighbors=7: 가장 가까운 7개의 이웃 데이터를 참조하여 분류하는 KNN 모델을 생성합니다.
knn = KNeighborsClassifier(n_neighbors=7)

# 학습 데이터(X_train)와 정답(y_train)을 사용하여 모델을 학습(fit)시킵니다.
knn.fit(X_train, y_train)
```

### 7. 시험 치기 (예측)

**1. 이유 (Why):** 숨겨뒀던 시험 문제(X_test)를 주고 정답을 맞춰보라고 합니다.

**2. 중요 포인트 (Key Point):** predict() 함수의 결과가 로봇이 써낸 답안지입니다.

**3. 주목할 점 (Focus):** 결과 배열에 0, 1, 2(꽃의 종류)가 나옵니다.


```python
# 예측
y_knn_pred = knn.predict(X_test)
print("예측값: ", y_knn_pred[:5])
```

### 8. 채점 하기

**1. 이유 (Why):** 로봇이 쓴 답과 실제 정답(y_test)을 비교해 점수를 매깁니다.

**2. 중요 포인트 (Key Point):** 1.0이면 100점입니다.

**3. 주목할 점 (Focus):** 점수가 너무 낮으면 공부 방법이나 재료를 바꿔야 합니다.


```python
# KNN 모델의 예측 성능을 평가합니다.
from sklearn.metrics import accuracy_score

# 테스트 데이터의 정답(y_test)과 예측값(y_knn_pred)을 비교하여 정확도(accuracy)를 계산합니다.
knn_acc = accuracy_score(y_test, y_knn_pred)
print("Accuracy: %.4f" % knn_acc)
```

```python
# SVM(Support Vector Machine) 분류기 모델을 임포트합니다.
from sklearn.svm import SVC

# kernel='rbf': 방사 기저 함수(RBF) 커널을 사용하는 SVC 모델을 생성합니다.
svc = SVC(kernel='rbf')

# 학습 데이터(X_train, y_train)로 모델을 학습시킵니다.
svc.fit(X_train, y_train)
```

### 7. 시험 치기 (예측)

**2. 중요 포인트 (Key Point):** predict() 함수의 결과가 로봇이 써낸 답안지입니다.

**3. 주목할 점 (Focus):** 결과 배열에 0, 1, 2(꽃의 종류)가 나옵니다.


### 8. 채점 하기

**1. 이유 (Why):** 로봇이 쓴 답과 실제 정답(y_test)을 비교해 점수를 매깁니다.

**2. 중요 포인트 (Key Point):** 1.0이면 100점입니다.

**3. 주목할 점 (Focus):** 점수가 너무 낮으면 공부 방법이나 재료를 바꿔야 합니다.


```python
# 예측
y_svc_pred = svc.predict(X_test)
print("예측값: ", y_svc_pred[:5])
# 성능 평가
svc_acc = accuracy_score(y_test, y_svc_pred)
print("Accuracy: %.4f" % svc_acc)
```

### 6. 모델 학습 (공부시키기)

**1. 이유 (Why):** 로봇(모델)을 불러와서 문제(X_train)와 답(y_train)을 주고 공부시킵니다(fit).

**2. 중요 포인트 (Key Point):** fit() 함수가 바로 '학습'을 시키는 명령어입니다.

**3. 주목할 점 (Focus):** 에러 없이 넘어가면 로봇이 규칙을 깨우친 것입니다.


```python
# 로지스틱 회귀(Logistic Regression) 분류기 모델을 임포트합니다.
from sklearn.linear_model import LogisticRegression

# 로지스틱 회귀 모델 객체를 생성합니다.
lrc = LogisticRegression()

# 학습 데이터(X_train, y_train)로 모델을 학습시킵니다.
lrc.fit(X_train, y_train)
```

### 5. 학습용/시험용 나누기

**1. 이유 (Why):** 로봇이 문제와 답을 몽땅 외우는(과적합) 것을 막기 위해, 공부할 문제(Train)와 시험 문제(Test)를 나눕니다.

**2. 중요 포인트 (Key Point):** test_size=0.2는 20%를 시험용으로 뺀다는 뜻입니다.

**3. 주목할 점 (Focus):** X_train(공부용), X_test(시험용)의 개수가 잘 나뉘었는지 보세요.


```python
# Overfitting(과적합)을 방지하고 일반화 성능을 높이기 위해 교차 검증을 수행합니다.
# 먼저, 훈련 데이터(X_train)를 다시 학습용(X_tr)과 검증용(X_val)으로 분리합니다.
# Hold-out 방식: 데이터셋을 학습/검증/테스트 세트로 나누어 사용하는 방법입니다.
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train,
                                            test_size=0.3,
                                            shuffle=True,
                                            random_state=20)

# 분리된 학습용 및 검증용 데이터셋의 크기를 확인합니다.
print(X_tr.shape, y_tr.shape)
print(X_val.shape, y_val.shape)
```

```python
# 테스트 데이터 예측 및 평가
# 검증이 완료된 모델을 사용하여 최종 테스트 데이터(X_test)에 대한 성능을 평가합니다.
y_test_pred = rfc.predict(X_test)
test_acc = accuracy_score(y_test, y_test_pred)
print("Test Accuracy: %.4f" % test_acc)
```

```python
# 데이터셋을 5개의 Fold로 분할하는 KFold 클래스 객체 생성
# n_splits=5: 데이터를 5개의 등분으로 나눕니다.
# shuffle=True, random_state=20: 데이터를 섞어서 나눕니다.
from sklearn.model_selection import KFold
kfold = KFold(n_splits=5, shuffle=True, random_state=20)

# 훈련용 데이터와 검증용 데이터의 행 인덱스를 각 Fold별로 구분하여 생성
num_fold = 1
for tr_idx, val_idx in kfold.split(X_train):
    print("%s Fold----------------------------------" % num_fold)
    print("훈련: ", len(tr_idx), tr_idx[:10])
    print("검증: ", len(val_idx), val_idx[:10])
    num_fold = num_fold + 1
```

### 8. 채점 하기

**1. 이유 (Why):** 로봇이 쓴 답과 실제 정답(y_test)을 비교해 점수를 매깁니다.

**2. 중요 포인트 (Key Point):** 1.0이면 100점입니다.

**3. 주목할 점 (Focus):** 점수가 너무 낮으면 공부 방법이나 재료를 바꿔야 합니다.


```python
# 훈련용 데이터와 검증용 데이터의 행 인덱스를 각 Fold별로 구분하여 생성
val_scores = []
num_fold = 1

for tr_idx, val_idx in kfold.split(X_train, y_train):
    # 훈련용 데이터와 검증용 데이터를 행 인덱스 기준으로 추출
    X_tr, X_val = X_train.iloc[tr_idx, :], X_train.iloc[val_idx, :]
    y_tr, y_val = y_train.iloc[tr_idx], y_train.iloc[val_idx]

    # 학습
    rfc = RandomForestClassifier(max_depth=5, random_state=20)
    rfc.fit(X_tr, y_tr)

    # 검증
    y_val_pred = rfc.predict(X_val)
    val_acc = accuracy_score(y_val, y_val_pred)
    print("%d Fold Accuracy: %.4f" % (num_fold, val_acc))
    val_scores.append(val_acc)
    num_fold += 1
```

```python
# 평균 Accuracy 계산
# 각 Fold별 정확도의 평균을 계산하여 모델의 일반화 성능을 평가합니다.
import numpy as np
mean_score = np.mean(val_scores)
print("평균 검증 Accuraccy: ", np.round(mean_score, 4))
```
