```python
import pandas as pd

# 행 인덱스/열 이름 지정하여, 데이터프레임 만들기
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']],
                   index=['준서', '예은'],
                   columns=['나이', '성별', '학교'])

# 데이터프레임 출력하기
print(df)
```

```python
#행 인덱스 확인하기
print(df.index)
```

```python
#열 이름 확인하기
print(df.columns)
```

```python
# 행 인덱스, 열 이름 변경하기
df.index=['학생1', '학생2']
df.columns=['연령', '남녀', '소속']

print(df)
```

```python
print(df.index)      #행 인덱스
```

```python
print(df.columns)    #열 이름
```
