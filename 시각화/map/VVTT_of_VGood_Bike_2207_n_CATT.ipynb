```python
import matplotlib.pyplot as plt

xs=df_g.index.to_list()			#dy_day(데이터 프레임)의 index(날짜, 시간)를 리스트로 저장
ys=df_g['이용시간(분)'].to_list()			#dy_day(테이터 프레임)의 volume 필드를 리스트로 저장

plt.figure(figsize=(10, 6))			#그래프 크기 지정
plt.xlabel('대여소명')				#그래프 x축 이름(label) 지정
plt.ylabel('이용시간(분)')				#그래프 y축 이름(label) 지정

plt.xticks(rotation=90)

plt.bar(xs, ys, width=0.6, color='gray')
```

```python
df_site = pd.read_csv('./3. 공공자전거 대여소 정보(22.06월 기준).csv',encoding='cp949')
df_site.head()
```

```python
df_g
```

```python
site_no = []

for i in df_g.index :
    a = int(i.split('.')[0])
    site_no.append(a)

site_no
```

```python
df_g['대여소번호'] = site_no
df_g
```

```python
df_g.info()
```

```python
df_site.info()
```

```python
df_map = pd.merge(df_g, df_site, how='left', on='대여소번호')
df_map
```

```python
df_map.info()
```

```python
df_map_f = df_map[['보관소(대여소)명','위도','경도']]
```

```python
df_map_f
```

```python
df_map_f = df_map_f.rename(columns={'보관소(대여소)명':'보관소명'})
```

```python
df_map_f
```

```python
import folium

pole_map = folium.Map(location=[37.54,126.99], tiles='OpenStreetMap',
                        zoom_start=12)

for name, lat, lng in zip(df_map_f.보관소명,df_map_f.위도, df_map_f.경도):
    folium.CircleMarker([lat, lng],
                        radius=8,         # 원의 반지름
                        color='Y',         # 원의 둘레 색상
                        fill=True,
                        fill_color='K',    # 원을 채우는 색
                        fill_opacity=0.7, # 투명도
                        popup=name
    ).add_to(pole_map)

pole_map

# 지도를 HTML 파일로 저장하기
pole_map.save('./bike_2206.html')
```
