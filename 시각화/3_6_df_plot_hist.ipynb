```python
import pandas as pd

# 데이터프레임 변환
df = pd.read_excel('./data/남북한발전전력량.xlsx')

# 남한, 북한 발전량 합계 데이터만 추출
df_ns = df.iloc[[0, 5], 3:]

# 행 인덱스 변경
df_ns.index = ['South','North']

# 발전량 데이터의 자료형을 문자열에서 정수형으로 변환
df_ns = df_ns.astype(int)

# 행, 열 전치하여 히스토그램 그리기
tdf_ns = df_ns.T
tdf_ns.plot(kind='hist');
```

```python
# 차트 서식 설정
tdf_ns.plot(kind='hist',
            alpha=0.5,          # 투명도 (0 ~ 1)
            figsize=(6, 2),     # 그림 크기 (가로, 세로)
            grid=True,          # 그리드 표시 여부
            color='skyblue',    # 차트 색
            edgecolor='black'); # 테두리 색
```

```python
# hist 메소드 활용
tdf_ns.hist(figsize=(6, 2));
```

```python
# 차트 서식 설정
tdf_ns.hist(bins=20, grid=False, figsize=(6, 2));
```
